## Database decompression
Remember to load the database from your computer. The file is dataset.zip or whatever name you gave it.

```python
import os 
import zipfile

local_zip = "dataset.zip" # Change the name to the one you have given it.
zip_ref = zipfile.ZipFile(local_zip, "r")
zip_ref.extractall("dataset")
zip_ref.close()
```

Make sure that the path we are defining is the same where you have the file

```python
path = "dataset/dataset"
content = os.listdir(path)
```

We visualize the content of the database

```python
content
```

## Distribuimos la base de datos en Training y Test
We define the path where we will store the distributed database as training and test

```python
path_general = "datasetFinal"
path_train = "{}/train".format(path_general)
path_test = "{}/test".format(path_general)

os.mkdir(path_general)
os.mkdir(path_train)
os.mkdir(path_test)
```

The factor is 70% train and 30% random test.

```python
import random
import shutil

# We initially store the training data.
train = 0.7
for nCount in range(int(len(content)*train)):
  random_choice_img = random.choice(content)
  random_choice_img_abs = "{}/{}".format(path,random_choice_img)
  target_img = "{}/{}".format(path_train, random_choice_img)
  shutil.copyfile(random_choice_img_abs, target_img)
  content.remove(random_choice_img)

```

```python
# We store the data left in the list as the test data.
for img in content:
  random_choice_img_abs = "{}/{}".format(path, img)
  target_img = "{}/{}".format(path_test, img)
  shutil.copyfile(random_choice_img_abs, target_img)
```
